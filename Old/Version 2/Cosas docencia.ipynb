```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 6, num=601)
x_0 = 0
v_0 = -5
a = 2.5
x = [x_0 + v_0 * t_i + 0.5 * a * t_i * t_i for t_i in t]
v = [v_0 + a * t_i for t_i in t]

R1_x = np.linspace(0, 0.5, num=101)
R1_y = [x_0 + v_0 * x for x in R1_x]
R2_x = np.linspace(1.5, 2.5, num=101)
R2_y = [-5 for x in R2_x]
R3_x = np.linspace(4, 4.5, num=101)
R3_y = [x_0 + - v_0 * x for x in R1_x]

plt.plot(t, x, label="Posición")
plt.plot(R1_x, R1_y, 'r--')
plt.plot(R2_x, R2_y, 'r--')
plt.plot(R3_x, R3_y, 'r--')

x1,x2,y1,y2 = plt.axis()
plt.axis((x1-0.3,x2+0.3,y1-1,y2+1))

plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')
plt.title('Posición en función del tiempo \n V_0 = -4m/s, a = 5m/s')
plt.grid(True)
plt.legend()

plt.axhline(0, color='red')
plt.axvline(0, color='red')
plt.arrow(1,0,0,-5, shape='full', lw=3, length_includes_head=True, head_width=0.3, color = 'green')
plt.arrow(3,-5,0,5, shape='full', lw=3, length_includes_head=True, head_width=0.3, color = 'blue')


plt.savefig("XvsTAnalisis.png")
plt.show()















plt.plot(t, v, label="Velocidad")

plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.title('Velocidad en función del tiempo \n V_0 = -4m/s, a = 5m/s')
plt.grid(True)
plt.legend()

plt.axhline(0, color='red')
plt.axvline(0, color='red')

x1,x2,y1,y2 = plt.axis()
plt.axis((x1-0.3,x2+0.3,y1-1,y2+1))

plt.fill_between(t[0:200], v[0:200], 0, color="none",hatch="X",edgecolor="g")
plt.fill_between(t[201:400], v[201:400], 0, color="none",hatch="X",edgecolor="b")

plt.savefig("VvsTAnalisis.png")
plt.show()
```
